```python
# Read with pandas
earthquake = pd.read_csv(earthquake, encoding="iso-8859-1", low_memory=False)
earthquake.head()
```

```python
earthquake.drop(['Depth Error','Magnitude Type', 'Magnitude Seismic Stations', 'Magnitude Error','Type','Depth Seismic Stations','Azimuthal Gap','Horizontal Distance', 'Horizontal Error','Root Mean Square' ], axis=1)
earthquake.tail()
```

```python
earthquake_df = earthquake.drop(['Depth Error', 'Depth Seismic Stations', 'Magnitude Seismic Stations', 'Type','Azimuthal Gap', 'Horizontal Error', 'Root Mean Square', 'Source','Location Source','Status','Magnitude Error','Magnitude Error','Horizontal Distance'], axis=1)
```

```python
# transform 'Date' from string to date format and add it to df
earth2 = pd.to_datetime(earthquake_df['Date'])
earthquake_df['Year_bydate'] = earth2.dt.year
earthquake_df.head()
```

```python
bydate = earthquake[['ID','Date', 'Latitude','Longitude']].copy()
bydate.head()
```

```python
# transform date string into date format and isolate the year per event
date_to_year = pd.to_datetime(bydate['Date'])
bydate['Year_bydate'] = date_to_year.dt.year

bydate.head()
```

```python
#df into csv to save it in local folder
bydate.to_csv(r'data_sets\date_location.csv')

print(bydate[:5].to_csv())
```

```python
engine = create_engine('sqlite:///earthquakesData.db')
```

```python
# save data into sqlite
earthquake_df.to_sql('earthquake', con=engine, if_exists='replace')
```
